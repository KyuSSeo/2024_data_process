```python
#HDF5

loaded = []

filename = "H5_data.h5"
df1.to_hdf(filename, key = 'df1', mode = 'w')
df2.to_hdf(filename, key = 'df2', mode = 'a')
df3.to_hdf(filename, key = 'df3', mode = 'a')
```

```python
with pd.HDFStore(filename, 'r')as store:
    for key in store.keys():
        loaded.append(store[key])
```
