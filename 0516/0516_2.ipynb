```python
img = cv2.imread("1.jpg");
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
```

```python
H,S,V = cv2.split(hsv)
```

```python
plt.figure(figsize = (12,4))
```

```python
plt.subplot(1,3,1)
plt.imshow(H,cmap = 'hsv')
plt.subplot(1,3,2)
plt.imshow(S,cmap = 'gray')
plt.subplot(1,3,3)
plt.imshow(V,cmap = 'gray')
plt.show
```

```python
img2 = cv2.imread("2.jpg");
img2_ = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize = (1,1))
plt.subplot(1,2,1)
plt.imshow(img2)
plt.title('BGR img')
plt.subplot(1,2,2)
plt.imshow(img2_)
plt.title('RGB img')
plt.show()
```

```python
R,G,B = cv2.split(img2_)
hsv2 = cv2.cvtColor(img2, cv2.COLOR_RGB2HSV)
H,S,V = cv2.split(hsv2)
```

```python
plt.figure(figsize = (1,1))
plt.subplot(1,3,1)
plt.imshow(R, cmap = 'Reds')
plt.title('R img')
plt.subplot(1,3,2)
plt.imshow(G, cmap = 'Greens')
plt.title('G img')
plt.subplot(1,3,3)
plt.imshow(B, cmap = 'Blues')
plt.title('B img')
plt.subplot(2,3,1)
plt.imshow(H, cmap = 'hsv')
plt.title('H img')
plt.subplot(2,3,2)
plt.imshow(S, cmap = 'gray')
plt.title('S img')
plt.subplot(2,3,3)
plt.imshow(V, cmap = 'gray')
plt.title('V img')
plt.show
```
